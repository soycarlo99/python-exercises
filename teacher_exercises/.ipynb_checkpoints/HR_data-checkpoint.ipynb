# DATA STORYTELLING Work accidents --------------------------------

```python
plt.figure(figsize=(12, 6))
dept_accident_rates = df.groupby('Department')['Work_accident'].mean().sort_values(ascending=False) * 100

ax = dept_accident_rates.plot(kind='bar', color='firebrick')
plt.title('Work Accident Rates by Department', fontsize=14, fontweight='bold')
plt.xlabel('Department', fontsize=12)
plt.ylabel('Accident Rate (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(dept_accident_rates):
    ax.text(i, v + 0.3, f'{v:.1f}%', ha='center', fontsize=10)

# Add company average line
company_avg = df['Work_accident'].mean() * 100
plt.axhline(y=company_avg, color='black', linestyle='--', 
            label=f'Company Average: {company_avg:.1f}%')

plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))

hour_bins = [150, 175, 200, 225, 250, 275, 300, 325]
labels = [f'{i}-{j}' for i, j in zip(hour_bins[:-1], hour_bins[1:])]
df['hour_group'] = pd.cut(df['average_montly_hours'], bins=hour_bins, labels=labels)

hours_accident = df.groupby('hour_group', observed=False)['Work_accident'].mean() * 100
hours_count = df.groupby('hour_group', observed=False).size()

fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:red'
ax1.set_xlabel('Monthly Working Hours', fontsize=12)
ax1.set_ylabel('Accident Rate (%)', color=color, fontsize=12)
line1 = ax1.plot(hours_accident.index, hours_accident.values, 'o-', color=color, linewidth=3, label='Accident Rate')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, hours_accident.max() * 1.2)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Number of Employees', color=color, fontsize=12)
bars = ax2.bar(np.arange(len(hours_count)), hours_count, alpha=0.3, color=color, 
               label='Employee Count')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_xticks(np.arange(len(hours_accident)))
ax2.set_xticklabels(hours_accident.index, rotation=45)

for i, v in enumerate(hours_accident):
    ax1.text(i, v + 0.2, f'{v:.1f}%', ha='center', color='darkred', fontweight='bold')

plt.title('Accident Rate vs. Monthly Working Hours', fontsize=14, fontweight='bold')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
```

```python
# Hours vs Evaluation Matrix with Accident Overlay
plt.figure(figsize=(12, 10))

plt.scatter(df[df['Work_accident']==0]['average_montly_hours'], 
            df[df['Work_accident']==0]['last_evaluation'], 
            alpha=0.5, c='blue', label='No Accident')
plt.scatter(df[df['Work_accident']==1]['average_montly_hours'], 
            df[df['Work_accident']==1]['last_evaluation'], 
            alpha=0.7, c='red', s=80, label='Had Accident')

hours_median = df['average_montly_hours'].median()
eval_median = df['last_evaluation'].median()
plt.axvline(x=hours_median, color='gray', linestyle='--')
plt.axhline(y=eval_median, color='gray', linestyle='--')

# Add quadrant labels
plt.text(df['average_montly_hours'].min() + 10, eval_median + 0.02, 
         'Low Hours, High Evaluation', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
plt.text(hours_median + 10, eval_median + 0.02, 
         'High Hours, High Evaluation', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
plt.text(df['average_montly_hours'].min() + 10, eval_median - 0.05, 
         'Low Hours, Low Evaluation', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
plt.text(hours_median + 10, eval_median - 0.05, 
         'High Hours, Low Evaluation', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.title('Work Hours vs. Performance Evaluation\nHighlighting Work Accidents', 
          fontsize=14, fontweight='bold')
plt.xlabel('Average Monthly Hours', fontsize=12)
plt.ylabel('Performance Evaluation', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
```

```python
# Impact of Work Accidents on Attrition
plt.figure(figsize=(12, 6))

# pivot table
pivot = pd.pivot_table(df, values='left', index='Department', 
                       columns=df['Work_accident'].map({0: 'No Accident', 1: 'Had Accident'}),
                       aggfunc='mean') * 100

# Sort by the difference (retention effect of accidents)
pivot['Difference'] = pivot['No Accident'] - pivot['Had Accident']
pivot = pivot.sort_values('Difference')

ax = pivot[['No Accident', 'Had Accident']].plot(kind='barh', figsize=(12, 8), 
                                               color=['tomato', 'forestgreen'])
plt.title('Attrition Rate: Employees With vs. Without Accidents', fontsize=14, fontweight='bold')
plt.xlabel('Attrition Rate (%)', fontsize=12)
plt.ylabel('Department', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend(title='Employee Status')

# Add data labels
for i, v in enumerate(pivot['No Accident']):
    ax.text(v + 0.5, i, f'{v:.1f}%', va='center')
for i, v in enumerate(pivot['Had Accident']):
    ax.text(v + 0.5, i, f'{v:.1f}%', va='center')
```

```python
plt.figure(figsize=(10, 6))

# Calculate accident rates by years at company
tenure_accident = df.groupby('time_spend_company')['Work_accident'].mean() * 100
tenure_count = df.groupby('time_spend_company').size()

fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:red'
ax1.set_xlabel('Years at Company', fontsize=12)
ax1.set_ylabel('Accident Rate (%)', color=color, fontsize=12)
bars = ax1.bar(tenure_accident.index, tenure_accident.values, color=color, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)

for i, v in enumerate(tenure_accident):
    ax1.text(i + 2, v + 0.3, f'{v:.1f}%', ha='center', color='darkred')

# Second axis for employee count
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Number of Employees', color=color, fontsize=12)
ax2.plot(tenure_count.index, tenure_count.values, 'o--', color=color, linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Work Accident Rate by Employee Tenure', fontsize=14, fontweight='bold')
plt.grid(axis='x')

# Add correlation information
corr = df['time_spend_company'].corr(df['Work_accident'])
plt.figtext(0.5, 0.01, f'Correlation between tenure and accident rate: {corr:.3f}',  
           ha='center', fontsize=11, bbox=dict(facecolor='white', alpha=0.8))
```

```python
%matplotlib inline
dept_metrics = df.groupby('Department')[['satisfaction_level', 'last_evaluation', 
                                       'Work_accident', 'left', 
                                       'promotion_last_5years']].mean()
                                       
# Normalize for radar chart, scale normalization 
for col in dept_metrics.columns:
    dept_metrics[col] = (dept_metrics[col] - dept_metrics[col].min()) / (dept_metrics[col].max() - dept_metrics[col].min())

# Setup radar chart
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, polar=True)

top_accident_depts = dept_metrics.sort_values('Work_accident', ascending=False).head(5).index

categories = dept_metrics.columns
N = len(categories)

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

# Plot for each department
for dept in top_accident_depts:
    values = dept_metrics.loc[dept].values.tolist()
    values += values[:1]
    
    # Plot values
    ax.plot(angles, values, linewidth=2, label=dept)
    ax.fill(angles, values, alpha=0.1)

# Set labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title('Department Comparison: Key Metrics', size=15)
plt.legend(loc='lower right')
```

```python
%matplotlib qt

# Select only numeric columns for correlation
numeric_cols = ['satisfaction_level', 'last_evaluation', 'number_project', 
                'average_montly_hours', 'time_spend_company', 'Work_accident', 
                'left', 'promotion_last_5years']

# Calculate correlation matrix
corr_matrix = df[numeric_cols].corr()

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x_pos, y_pos = np.meshgrid(
    range(len(numeric_cols)), 
    range(len(numeric_cols))
)
x_pos = x_pos.flatten()
y_pos = y_pos.flatten()
z_pos = np.zeros_like(x_pos)
dx = dy = 0.7 * np.ones_like(z_pos)
dz = corr_matrix.values.flatten()

colors = plt.cm.coolwarm(0.5 * (dz + 1))

bars = ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color=colors, alpha=0.8)

# Add a color bar
color_map = plt.cm.ScalarMappable(cmap='coolwarm')
color_map.set_array(dz)
cbar = fig.colorbar(color_map, ax=ax, pad=0.1, shrink=0.7, label='Correlation')

# Labels
ax.set_xticks(np.arange(len(numeric_cols)) + 0.35)
ax.set_yticks(np.arange(len(numeric_cols)) + 0.35)
ax.set_xticklabels(numeric_cols, rotation=45, ha='right')
ax.set_yticklabels(numeric_cols, rotation=45, ha='right')
ax.set_zlabel('Correlation Strength')

ax.set_title('3D Correlation Matrix', fontsize=14)

plt.show()
```
